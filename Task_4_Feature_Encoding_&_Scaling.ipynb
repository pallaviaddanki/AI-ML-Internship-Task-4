```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python

columns = [
    "age", "workclass", "fnlwgt", "education", "education-num",
    "marital-status", "occupation", "relationship", "race", "sex",
    "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"
]

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"

df = pd.read_csv(
    url,
    header=None,
    names=columns,
    sep=", ",
    engine="python"
)

print("Dataset loaded successfully!")
print(df.head())
```

```python

df.replace("?", np.nan, inplace=True)
df.dropna(inplace=True)

print("\nMissing values handled!")
print(df.isnull().sum())
```

```python
numerical_features = [
    "age", "fnlwgt", "education-num",
    "capital-gain", "capital-loss", "hours-per-week"
]

categorical_features = [
    "workclass", "education", "marital-status",
    "occupation", "relationship", "race",
    "sex", "native-country"
]

target = "income"

print("\nNumerical Features:", numerical_features)
print("Categorical Features:", categorical_features)
```

```python
df_encoded = pd.get_dummies(
    df,
    columns=categorical_features,
    drop_first=True
)

print("\nOne-Hot Encoding completed!")
print(df_encoded.shape)
```

```python

scaler = StandardScaler()
df_encoded[numerical_features] = scaler.fit_transform(
    df_encoded[numerical_features]
)

print("\nNumerical features scaled!")
```

```python
print("\nBefore Scaling (Sample):")
print(df[numerical_features].head())

print("\nAfter Scaling (Sample):")
print(df_encoded[numerical_features].head())
```

```python
df_encoded.to_csv("adult_preprocessed.csv", index=False)

print("\nPreprocessed dataset saved as 'adult_preprocessed.csv'")
```
